```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from PyPDF2 import PdfReader
```

```python
from google.colab import files
uploaded=files.upload()
pdf_path=list(uploaded.keys())[0]
print(f"Uploaded file:{"~/Downloads/Factor Analysis_Theory.pdf"}")
```

```python
reader=PdfReader("Factor Analysis_Theory.pdf")
text=""
for page in reader.pages:
  text+=page.extract_text()
print(text[:1000])
```

```python
text_splitter=RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len
)
chunks=text_splitter.split_text(text)
print(f"chunks created:{len(chunks)}")
print(chunks[0][:500])
```

```python
!pip install -U langchain_community
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
model_name="sentence-transformers/all-MiniLM-L6-v2"
embeddings=HuggingFaceEmbeddings(model_name=model_name)
```

```python
from langchain.vectorstores import Chroma
```

```python
vectorstore=Chroma.from_texts(chunks,embedding=embeddings,persist_directory="./chroma_db")
```

```python
query="What is the underlying concept of PCA?"
results=vectorstore.similarity_search(query,k=3)
for i,r in enumerate(results):
  print(f"\n---Result{i+1}---n")
  print(r.page_content[:500])
```

```python
vectorstore.persist()
```
